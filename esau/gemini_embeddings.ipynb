```python
def embed_resumes():
    print("🔍 Loading resumes...")
    df_resumes = pd.read_csv("../resume_queries.csv")

    print("🔨 Generating embeddings for resumes...")
    tqdm.pandas(desc="Embedding resumes")
    df_resumes["embedding"] = df_resumes["Resume_str"].progress_apply(lambda x: get_embedding(str(x)))

    # Keep only ID, resume_str, and embedding
    df_output = df_resumes[["ID", "Resume_str", "embedding"]]

    df_output.to_csv("gemini_embeddings_resumes.csv", index=False)
    print("✅ Saved resume embeddings to 'gemini_embeddings_resumes.csv'")
embed_resumes()
```
